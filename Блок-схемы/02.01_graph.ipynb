```python
import networkx as nx

g = nx.DiGraph()

g.add_nodes_from([1,2,3,4,5])
g.add_edges_from([(1,4), (4,1), (4,3), (2,4), (5,1)])

g = nx.petersen_graph()

res = nx.dfs_predecessors(g, source=5)
print(res)
res1 = list(nx.dfs_preorder_nodes(g, source=5))
print(res1)

nx.draw(g, with_labels=True)


```
